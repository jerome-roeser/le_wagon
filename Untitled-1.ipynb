### No need for a `value_counts()` with `seaborn`, we can just use `Countplot`

```python
_ = sns.countplot(df, x= df['neighbourhood_group'])
```

but the plot is not sorted. This is done by the method `value_counts`

```python
df['neighbourhood_group'].value_counts()
```

so an easy way is to store this ordered values with their `index`

```python
df['neighbourhood_group'].value_counts().index
```

```python
order = df['neighbourhood_group'].value_counts().index
sns.countplot(df, x= df['neighbourhood_group'], order=order)
sns.set_style
```

## What is the distribution of flat prices?

```python
_ = sns.displot(df['price'])

```

```python
df[df['price'] > 500].shape
```

```python
affordable_df = df[df['price'] <= 500]
sns.displot(affordable_df['price'])
```

```python
affordable_df['price'].describe()
```

## What is the distribution of flat prices based on the neighboorhood groups?

```python
plt.figure(figsize=(15,8))
sns.set_palette('deep')
sns.set_
_ = sns.violinplot(affordable_df, x = 'neighbourhood_group', y = 'price')
affordable_df.columns
```

## Can we plot the listings on a map?

trick: use `latitude` and `longitude`

```python
affordable_df.plot(
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha= 0.8,
    figsize=(12,8)
)
```
